```python
def SIR_Step(S,I,R,b,k):
    S+=-b*S*I
    R+=k*I
    I+=b*S*I-k*I
    if S<=0: S=0
    if I<=0: I=0
    if R<=0: R=0
    
    return [S,I,R]

#Initiate Cities
C1=[20000,10,3,.0003,.0002]
C2=[25000,15,3,.0003,.0002]
C3=[14000,20,3,.0003,.0002]
C4=[10000,0,3,.0003,.0002]

#Edges with probabilities
EdgesC1=[[C2,.003],[C4,.004]]
EdgesC2=[[C3,.001],[C1,.005]]
EdgesC3=[[C4,.005],[C2,.004]]
EdgesC4=[[C1,.002],[C3,.001]]

Cities=[[C1,EdgesC1],[C2,EdgesC2],[C3,EdgesC3],[C4,EdgesC4]]

def Step(Cities):
    #First each city steps individually
    for i in Cities:
        S=i[0][0]
        I=i[0][1]
        R=i[0][2]
        b=i[0][3]
        k=i[0][4]
        SIR = SIR_Step(S,I,R,b,k)
        S = SIR[0]
        I= SIR[1]
        R = SIR[2] 
        #update values in the cities
        i[0][0]=S
        i[0][1]=I
        i[0][2]=R
        
        
        #Now people move between cities
        for j in i[1]:
            dS=S*j[1]
            i[0][0]+=-dS
            j[0][0]+=dS
            
            dI=I*j[1]
            i[0][1]+=-dI
            j[0][1]+=dI
            
            dR=I*j[1]
            i[0][2]+=-dI
            j[0][2]+=dI
    return Cities

print Step(Cities)
print''
print Step(Cities)
print''
print Step(Cities)
print ''
print Step(Cities)
print ''
print Step(Cities)
            
        
```
