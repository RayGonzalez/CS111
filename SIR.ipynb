```python
def SIR_Step(S,I,R,b,k):
    S+=-b*S*I
    R+=k*I
    I+=b*S*I-k*I
    if S<=0: S=0
    if I<=0: I=0
    if R<=0: R=0
    
    return [S,I,R]


S=10000
I=2
R=0
b=.0003
k=.0002


for i in range(0,1000):
    SIR = SIR_Step(S,I,R,b,k)
    S = SIR[0]
    I= SIR[1]
    R = SIR[2]
    print "S=",S
    print 'I=',I
    print 'R=', R
    print ''


    
    
```
